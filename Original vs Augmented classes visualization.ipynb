# Original Dataset

```python
import os
import matplotlib.pyplot as plt
from matplotlib.image import imread
plt.rcParams['figure.dpi'] = 500
plt.rcParams['savefig.dpi'] =500
```

```python
data_dir = 'D:\Fall 2022\CV project\dataset\Alzheimer'
train_path = data_dir+'\\train'
os.listdir(train_path)
```

```python
train_paths = []
train_labels = []

for label in os.listdir(train_path):
    for image in os.listdir(train_path+'\\'+label):
        train_paths.append(train_path+label+'/'+image)
        train_labels.append(label)
```

```python
plt.figure(figsize=(14,6))
colors = ['#4285f4', '#ea4335', '#fbbc05', '#34a853']
plt.rcParams.update({'font.size': 14})
plt.pie([len([x for x in train_labels if x=='MildDemented']),
         len([x for x in train_labels if x=='ModerateDemented']),
         len([x for x in train_labels if x=='NonDemented']),
         len([x for x in train_labels if x=='VeryMildDemented'])],
        labels=['MildDemented', 'ModerateDemented', 'NonDemented', 'VeryMildDemented'],
        colors=colors, autopct='%.1f%%', explode=(0.025,0.025,0.025,0.025),
        );
```
